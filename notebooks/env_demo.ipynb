# Render gym in jupyter notebook

This notebook demonstrates how a gym environment run might be displayed in a jupyter 
notebook.

```python
import gymnasium as gym 
import matplotlib.pyplot as plt
from templateRL import PPOAgent
from IPython.display import clear_output
```

```python
env = gym.make("LunarLander-v3", render_mode='rgb_array')
```

```python
env.reset()
plt.imshow(env.render())
plt.axis('off')
```

```python
agent = PPOAgent.load("../ppo.agent")
```

```python
rollout = agent.play_episode(env)
```

```python
env.close()
```

```python
env = gym.make("LunarLander-v3", render_mode='rgb_array')
original_step_func = env.step
def new_step(action):
    r = original_step_func(action)
    clear_output(True)
    plt.imshow(env.render())
    plt.axis('off')
    plt.pause(0.0001)
    return r
env.step = new_step
```

```python
rollout = agent.play_episode(env)
```
